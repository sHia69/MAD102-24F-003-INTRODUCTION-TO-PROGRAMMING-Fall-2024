B. Write a Python program that converts a binary number received as user input into its hexadecimal equivalent.

Create function `binaryHex()`:

- The function should take a binary string as input.
- It should convert the binary string to a decimal number, then convert the decimal number to a hexadecimal string.
- The function should display the binary number entered by the user and its hexadecimal equivalent.
- Ensure hexadecimal output is in uppercase (e.g., "A", "2F").

Program Flow:

- Prompt the user to input a binary number. Validate the input to ensure it only contains 0s and 1s.
- If the input is invalid, prompt the user to re-enter a valid binary number.
- Format the output & Input Prompt as given below: 

Sample Input:

```
Enter a binary number: 1122 
```

Sample Output:

```
Invalid input. Please enter a binary number with only 0's and 1's.
```

Sample Input:

```
Enter a binary number: 1011
```

Sample Output:

```
Binary: 1011 ---> Hexadecimal: B
```

````python
def countItems(nested_list):
    """
    Function to count the number of elements in each inner list of a nested list,
    and print each inner list with its respective count of elements.
    """
    # Loop through the nested list and get inner list and its index
    for index, inner_list in enumerate(nested_list):
        # Count the number of elements in the inner list
        count = len(inner_list)
        # Print the inner list and its count
        print(f"Inner list {index+1}: {inner_list} - Count: {count}")


nested_list = [[1, 2, 3], ['apple', 'banana']]
countItems(nested_list)
````

````python
def binaryHex():
    """
    This function takes a binary string as input, converts it to decimal, 
    and then converts the decimal number to hexadecimal. It displays the 
    binary input and its hexadecimal equivalent in uppercase.
    """
    
    # Get the binary numbermas input
    binary_str = input("Enter a binary number: ")
    
    # Check if the input contains only '0' and '1's
    while not all(digit in '01' for digit in binary_str):
        # If the input is invalid, ask the user to enter the binary number
        print("Invalid input. Please enter a binary number with only 0's and 1's.")
        binary_str = input("Enter a binary number: ")
    
    # Convert the binary string to a decimal integer
    decimal_num = int(binary_str, 2)
    
    # Convert the decimal integer to a hexadecimal string
    hex_num = hex(decimal_num)[2:].upper()  
  
    print(f"Binary: {binary_str} ---> Hexadecimal: {hex_num}")


binaryHex()
````

B. Write a Python program that performs operations on a dictionary where:

- One key contains a list of numbers.
- Another key contains a list of strings.

The program should provide the following functionality using functions:

**Find the Maximum Element by Length/Value:**

Create a function that takes the dictionary as input and finds the maximum element:
- The maximum number from the list under the key 'numbers'.
- The string with the maximum length from the list under the key 'strings'.

**Find the Minimum Element by Length/Value:**

Create a function that takes the dictionary as input and finds the minimum element:
- The minimum number from the list under the key 'numbers'.
- The string with the minimum length from the list under the key 'strings'.

**Program Flow:**

The program should provide the following menu options for the user:
- Option 1: Find the maximum element by value (for numbers) and by length (for strings) and display them as output.
- Option 2: Find the minimum element by value (for numbers) and by length (for strings) and display them as output.
- Option 3: Exit the program.

Format of the output must be similar to sample input & output: Given dictionary is `{"numbers":[2,5,100], "strings":['Apple', 'banana','kiwi']}`

**Sample Output & Input:**

```
------------------------------

2. Find the minimum element by value/length

3. Exit Enter your choice (1/2/3): 1

Maximum string by length: banana

------------------------------

1. Find the maximum element by value/length

2. Find the minimum element by value/length

Enter your choice (1/2/3): 2

Minimum number: 5

Minimum string by length: kiwi

1. Find the maximum element by value/length

2. Find the minimum element by value/length

```

````python
def create_dictionary():
    """
    Function to create a dictionary from a list of words entered by the user.
    Each key is the first letter of a word, and the value is a list of words
    starting with that letter.
    """
    # Get the input from the user
    user_input = input("Enter words separated by commas and spaces: ")

    # Split the input into individual words, strip whitespace, and convert them to lowercase
    words = [word.strip().lower() for word in user_input.split(',')]

    # Create the empty dictionary
    word_dict = {}

    for word in words:
        # Get the first letter of each word
        first_letter = word[0]

        # Add the word to the dictionary using the first letter as key
        if first_letter in word_dict:
            word_dict[first_letter].append(word)
        else:
            word_dict[first_letter] = [word]

    return word_dict

# Create the dictionary using the function "create_dictionary"
word_dict = create_dictionary()

# Prompt user for a letter to retrieve words starting with it
letter = input("\nEnter the first letter of words to display: ").strip().lower()

# Check if the letter exists in the dictionary
if letter in word_dict:
    print(f"\nWords starting with '{letter}'")
    print("-" * 30)
    for word in word_dict[letter]:
        # Capitalize the first letter of each word before printing
        print(f"          {word.capitalize()}")
else:
    print(f"\nNo words found starting with '{letter}'.")
````

````python
def maxByLengthorValue(data_dict):
    """
    Function to find the maximum element by value (for numbers)
    and by length (for strings).
    """
    # Find the max number 
    max_num = max(data_dict['numbers'])
    
    # Find the string with the max length 
    max_str = max(data_dict['strings'], key=len)
    
    return max_num, max_str

def minByLengthorValue(data_dict):
    """
    Function to find the minimum element by value (for numbers)
    and by length (for strings).
    """
    # Find the min number 
    min_num = min(data_dict['numbers'])
    
    # Find the string with the min length 
    min_str = min(data_dict['strings'], key=len)
    
    return min_num, min_str

# Dictionary containing two keys: 'numbers' and 'strings'
data_dict = {
    'numbers': [10, 5, 25, 50, 100],
    'strings': ["apple", "banana", "kiwi", "grape", "orange"]
}

while True:
    # Display menu options
    print("-"*30)
    print("\nSelect an option")
    print("-"*30)
    print("\t1. Find the maximum element by value/length")
    print("\t2. Find the minimum element by value/length")
    print("\t3. Exit")
    
    # Get the user's choice
    choice = input("Enter your choice (1/2/3): ")
    
    if choice == '1':
        # Call function to find the maximum element
        max_num, max_str = maxByLengthorValue(data_dict)
        print(f"Maximum number: {max_num}")
        print(f"Maximum string by length: {max_str}")
    
    elif choice == '2':
        # Call function to find the minimum element
        min_num, min_str = minByLengthorValue(data_dict)
        print(f"Minimum number: {min_num}")
        print(f"Minimum string by length: {min_str}")
    
    elif choice == '3':
        # Exit the program
        print("Exiting the program.")
        print("-"*30)
        break
    
    else:
        print("Invalid choice. Please enter 1, 2, or 3.")
````
