## Question 13
Answer any one of the questions below,

A. Write a python program that checks if a given year is a leap year or not

If the year is divisible by 400, print `<year> is a leap year.`  
If the year is divisible by 4 but not divisible by 100, print `<year> is a leap year.`  
Otherwise, print `<year> is not a leap year.`  
Note: you must get year as user input. You must format the output as shown in example output.

Example Output:

```
Enter a year: 2020
------------------------------------------------
2020 is a leap year.
```

B. Write a python program that checks if a given number is even or not

If the number is divisible by 2, print `<number> is an even number.`  
If the number is not divisible by 2 and leaves remainder 1, print `<number> is an odd number.`  
Non-integer numbers are neither even nor odd, such as 1/2, 0.88, or infinity. Print `<number> is neither even nor odd.`  
Note: you must get number as user input and convert it to float. You must format the output as shown in example output.

Example Output:

```
Enter a number: 100.0
------------------------------------------------
100.0 is an even number.
```

````python
# Question A: Check if a given year is a leap year or not
try:
    year = int(input("Enter a year: "))
    print("------------------------------------------------")
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        print(f"{year} is a leap year.")
    else:
        print(f"{year} is not a leap year.")
except ValueError:
    print("Invalid input. Please enter a valid year.")
          
````

````python
# Question B: Check if a given number is even or odd
try:
    number = float(input("Enter a number: "))
    print("------------------------------------------------")
    if number.is_integer():
        if number % 2 == 0:
            print(f"{number} is an even number.")
        else:
            print(f"{number} is an odd number.")
    else:
        print(f"{number} is neither even nor odd.")
except ValueError:
    print("Invalid input. Please enter a valid number.")
````

Questions to be asked in the quiz:

B. Develop a Python program to register k players for a gaming tournament and generates Player Tag. Program should collect information from user such as Player Name, Game Type, and Age and generate a Player Tag in the format: “GameType-Age-First 2 letters of Player Name-Random Number between 500 and 999”. Print the Player Tag for each participant.

Output prompts and formatting must be as mentioned below:

```
Enter the number of players: 1
----------------------------------------------------
Player 1:
----------------------------------------------------
Enter Player Name: David
Enter Game Type: Chess
Enter Age: 25
Player Tag: Chess-25-Da-783
```

````python
points = 0

# Question 1
answer1 = input("Question 1: What is the capital of France? ").strip().lower()
if answer1 == "paris":
          points += 20
          print("Correct! You have earned 20 points.")
else:
          print("Incorrect! You have been eliminated from the game.")
          print(f"Total points: {points}")
          exit()

# Question 2
answer2 = input("Question 2: What is 5 + 7? ").strip()
if answer2 == "12":
          points += 30
          print("Correct! You have earned 30 points.")
else:
          print("Incorrect! You have been eliminated from the game.")
          print(f"Total points: {points}")
          exit()

# Question 3
answer3 = input("Question 3: What is the largest planet in our solar system? ").strip().lower()
if answer3 == "jupiter":
          points += 50
          print("Correct! You have earned 50 points.")
else:
          print("Incorrect! You have been eliminated from the game.")
          print(f"Total points: {points}")
          exit()

print(f"Congratulations! You have completed the quiz with a total of {points} points.")
````

````python
import random

# Get the number of players
k = int(input("Enter the number of players: "))
print("----------------------------------------------------")

# Register each player and generate Player Tag
for i in range(1, k + 1):
          print(f"Player {i}:")
          print("----------------------------------------------------")
          player_name = input("Enter Player Name: ").strip()
          game_type = input("Enter Game Type: ").strip()
          age = int(input("Enter Age: ").strip())
          
          # Generate Player Tag
          random_number = random.randint(500, 999)
          player_tag = f"{game_type}-{age}-{player_name[:2]}-{random_number}"
          
          # Print Player Tag
          print(f"Player Tag: {player_tag}")
          print("----------------------------------------------------")
````
